```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Given data
data = np.array([[0.2, 0.3], [0.4, 0.5], [0.7, 0.9], [0.2, 0.8], [0.8, 0.6], [0.5, 0.7], [0.1, 0.4], [0.6, 0.2], [0.9, 0.5]])

# Fit k-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)  # Explicitly set n_init
kmeans.fit(data)

# Predict the cluster for each data point
cluster_assignments = kmeans.predict(data)

# Print the cluster assignments for each data point
for i, point in enumerate(data):
    print(f"Data point {point} belongs to Cluster {cluster_assignments[i]}")

# Predict the cluster for a new case with VAR1=0.906 and VAR2=0.606
new_case = np.array([[0.906, 0.606]])
predicted_cluster = kmeans.predict(new_case)

# Display the predicted cluster
print(f"\nThe predicted cluster for VAR1=0.906 and VAR2=0.606 is: {predicted_cluster[0]}")

# Plotting the clusters
plt.scatter(data[:, 0], data[:, 1], c=cluster_assignments, cmap='viridis', marker='o', edgecolor='k', s=100, label='Data Points')
plt.scatter(new_case[:, 0], new_case[:, 1], c='red', marker='x', s=200, label='New Data Point')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='blue', marker='^', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('VAR1')
plt.ylabel('VAR2')
plt.legend()
plt.show()
```
