```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('enjoysport.csv')
data = pd.DataFrame(dataset)

print("The dataset is:")
print(data)

concepts = np.array(data.iloc[:, :-1])
targets = np.array(data.iloc[:, -1])

print("\nConcepts:")
print(concepts)
print("\nTargets:")
print(targets)


def learn(concepts, targets):
    print("\Initializing the specific and general hypothesis")
    specific_h = concepts[0]
    general_h = [['?' for j in specific_h] for j in specific_h]
    
    print(specific_h)
    print(general_h)
    
    for i, h in enumerate(concepts):
        if targets[i] == 'yes':
            for j in range(len(specific_h)):
                if h[j] != specific_h[j]:
                    specific_h[j] = "?"
                    general_h[j][j] = "?"
        
        else:
            for j in range(len(concepts[i])):
                if h[j] != specific_h[j]:
                    general_h[j][j] = specific_h[j]
        
        print("\nSteps in Candidate Elimination Algorithm", i+1)
        print(specific_h)
        print(general_h)
    
    return specific_h, general_h

final_s, final_g = learn(concepts, targets)
print("\nFinal Specefic Hypothesis")
print(final_s)
print("\nFinal General Hypothesis")
print(final_g)
```
